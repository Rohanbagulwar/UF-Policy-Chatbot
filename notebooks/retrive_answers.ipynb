```python
from main import RetrievalPipeline
```

```python
pipeline = RetrievalPipeline()
pipeline.initialize()
```

```python


import pandas as pd

# 1. Load your Excel
df = pd.read_excel("test_Q_A .xlsx")

# df["gpt_oss"] = ""
df['gemma-3-27b-it']=''
count = 0
for index, row in df.iterrows():
    question = row["Question"]
    try:
        result = pipeline.query(
            question=question,
            n_results=3
        )
        answer = result['answer']
        print("answer got for:", answer)

        count += 1
    except Exception as e:
        answer = f"Error: {e}"

    df.at[index, "gemma-3-27b-it"] = answer
# 4. Save back to Excel
df.to_excel("test_Q_A_with_gemma_answers.xlsx", index=False)

print("‚úì Completed ‚Äî File saved as test_Q_A_with_llama_answers.xlsx")
```
