```python
import json
import asyncio
import aiohttp
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime
import re
from langchain_text_splitters import RecursiveCharacterTextSplitter

async def fetch_page(session, url):
    """Async fetch a single page."""
    try:
        headers = {'User-Agent': 'Mozilla/5.0'}
        async with session.get(url, headers=headers, timeout=30) as response:
            return await response.text()
    except Exception as e:
        print(f"Error fetching {url}: {e}")
        return None

```

```python
def extract_content(html):
    """Extract clean text content from HTML."""
    if not html:
        return ""
    
    soup = BeautifulSoup(html, 'html.parser')
    
    # Remove unwanted elements
    for element in soup(['script', 'style', 'nav', 'header', 'footer', 'aside']):
        element.decompose()
    
    # Get text
    text = soup.get_text(separator=' ', strip=True)
    # Clean multiple spaces and newlines
    text = re.sub(r'\s+', ' ', text)
    return text.strip()
```

```python

def split_into_statements(text,statements_per_chunk=10, chunk_size=1000, chunk_overlap=100):
    """
    Split text using LangChain's RecursiveCharacterTextSplitter.
    
    Args:
        text: The text to split
        statements_per_chunk: Approximate number of statements (used to calculate chunk_size)
        chunk_size: Maximum size of each chunk in characters
        chunk_overlap: Number of characters to overlap between chunks
    
    Returns:
        List of text chunks
    """
    if not text:
        return []
    
    # Initialize LangChain text splitter
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=chunk_size,
        chunk_overlap=chunk_overlap,
        length_function=len,
        separators=["\n\n", "\n", ". ", " ", ""]
    )
    
    # Split the text
    chunks = text_splitter.split_text(text)
    
    return chunks
```

```python
async def process_policy(session, policy_data, chunk_size=1000, chunk_overlap=100):
    """Process a single policy: fetch, parse, and chunk."""
    url = policy_data['link']
    print(f"Fetching: {policy_data['title']}")
    
    html = await fetch_page(session, url)
    content = extract_content(html)
    
    # Split content into chunks
    chunks = split_into_statements(content, chunk_size=chunk_size, chunk_overlap=chunk_overlap)
    
    # Create rows for each chunk
    rows = []
    for idx, chunk in enumerate(chunks, 1):
        rows.append({
            'title': policy_data['title'],
            'category': policy_data['category'],
            'type': policy_data['type'],
            'link': url,
            'chunk_number': idx,
            # 'total_chunks': len(chunks),
            'content': chunk
        })
    
    print(f"✓ {policy_data['title']}: {len(chunks)} chunks created")
    return rows
```

```python
async def fetch_all_policies(policies, chunk_size=1000, chunk_overlap=100,concurrent_limit=20):
    """Fetch all policies concurrently with a limit."""
    connector = aiohttp.TCPConnector(limit=concurrent_limit)
    async with aiohttp.ClientSession(connector=connector) as session:
        tasks = [process_policy(session, policy, chunk_size, chunk_overlap) for policy in policies]
        results = await asyncio.gather(*tasks, return_exceptions=True)
        
        # Flatten results and handle exceptions
        all_rows = []
        for result in results:
            if isinstance(result, Exception):
                print(f"Error processing policy: {result}")
            elif result:
                all_rows.extend(result)
        
        return all_rows
```

```python
def save_to_excel(data, output_file="policies_data.xlsx"):
    """Save data to Excel file."""
    try:
        df = pd.DataFrame(data)
        
        # Reorder columns
        column_order = ['title', 'category', 'type', 'link', 'chunk_number', 'content']
        df = df[column_order]
        
        # Create Excel writer with formatting
        with pd.ExcelWriter(output_file, engine='openpyxl') as writer:
            df.to_excel(writer, sheet_name='Policies', index=False)
            
            # Get workbook and worksheet
            workbook = writer.book
            worksheet = writer.sheets['Policies']
            
            # Adjust column widths
            worksheet.column_dimensions['A'].width = 30  # title
            worksheet.column_dimensions['B'].width = 20  # category
            worksheet.column_dimensions['C'].width = 15  # type
            worksheet.column_dimensions['D'].width = 50  # link
            worksheet.column_dimensions['E'].width = 12  # chunk_number
            # worksheet.column_dimensions['F'].width = 12  # total_chunks
            worksheet.column_dimensions['F'].width = 80  # content
            
            # Wrap text in content column
            from openpyxl.styles import Alignment
            for row in worksheet.iter_rows(min_row=2, max_row=worksheet.max_row, min_col=6, max_col=6):
                for cell in row:
                    cell.alignment = Alignment(wrap_text=True, vertical='top')
        
        print(f"\n✓ Data saved to {output_file}")
        print(f"✓ Total rows: {len(df)}")
        return df
    except Exception as e:
        print(f"Error saving to Excel: {e}")
        return None
```

```python
async def main(json_file, output_excel="policies_data.xlsx", chunk_size=1000, chunk_overlap=100,concurrent_limit=20):
    """Main function to orchestrate the entire process."""
    print("="*80)
    print("POLICY SCRAPER WITH ASYNC FETCH AND EXCEL EXPORT")
    print("="*80)
    
    # Load policies from JSON
    policies = load_policies_from_json(json_file)
    
    if not policies:
        print("No policies to process!")
        return
    
    print(f"\nProcessing {len(policies)} policies with {concurrent_limit} concurrent connections...")
    print(f"Chunking content into {chunk_size} characters per chunk with {chunk_overlap} characters overlap\n")
    
    # Fetch and process all policies
    start_time = datetime.now()
    all_data = await fetch_all_policies(policies, chunk_size,chunk_overlap,concurrent_limit)
    end_time = datetime.now()
    
    # Save to Excel
    if all_data:
        df = save_to_excel(all_data, output_excel)
        
        print(f"\n{'='*80}")
        print(f"✓ Processing complete!")
        print(f"✓ Time taken: {(end_time - start_time).total_seconds():.2f} seconds")
        print(f"✓ Policies processed: {len(policies)}")
        print(f"✓ Total rows created: {len(all_data)}")
        print(f"{'='*80}")
        
        return df
    else:
        print("No data to save!")
        return None
```

```python
async def fetch_page_retry(session, url, max_retries=3, delay=2):
    """Async fetch a single page with retry logic."""
    for attempt in range(max_retries):
        try:
            headers = {
                'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36',
                'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
                'Accept-Language': 'en-US,en;q=0.5',
            }
            async with session.get(url, headers=headers, timeout=30, ssl=False) as response:
                if response.status == 200:
                    return await response.text()
                else:
                    print(f"  Attempt {attempt+1}: Status {response.status}")
        except Exception as e:
            print(f"  Attempt {attempt+1} failed: {e}")
            if attempt < max_retries - 1:
                await asyncio.sleep(delay * (attempt + 1))  # Exponential backoff
    
    print(f"   All {max_retries} attempts failed for {url}")
    return None

def extract_content(html):
    """Extract clean text content from HTML."""
    if not html:
        return ""
    
    soup = BeautifulSoup(html, 'html.parser')
    
    # Remove unwanted elements
    for element in soup(['script', 'style', 'nav', 'header', 'footer', 'aside', 'iframe', 'noscript']):
        element.decompose()
    
    # Try to find main content area
    main_content = (
        soup.find('main') or 
        soup.find('article') or 
        soup.find('div', class_=re.compile(r'content|main|body', re.I)) or
        soup.find('body')
    )
    
    if main_content:
        text = main_content.get_text(separator=' ', strip=True)
    else:
        text = soup.get_text(separator=' ', strip=True)
    
    # Clean multiple spaces and newlines
    text = re.sub(r'\s+', ' ', text)
    return text.strip()

def split_text_langchain(text, chunk_size=1000, chunk_overlap=100):
    """Split text using LangChain's RecursiveCharacterTextSplitter."""
    if not text:
        return []
    
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=chunk_size,
        chunk_overlap=chunk_overlap,
        length_function=len,
        separators=["\n\n", "\n", ". ", " ", ""]
    )
    
    chunks = text_splitter.split_text(text)
    return chunks

async def process_single_policy(session, policy_data, chunk_size=1000, chunk_overlap=100):
    """Process a single policy with retry logic."""
    url = policy_data['link']
    title = policy_data['title']
    
    print(f"\n Processing: {title}")
    print(f"   URL: {url}")
    
    html = await fetch_page_retry(session, url, max_retries=3, delay=2)
    
    if not html:
        print(f"    Failed to fetch content")
        return []
    
    content = extract_content(html)
    
    if not content or len(content) < 50:
        print(f"    No meaningful content extracted (length: {len(content)})")
        return []
    
    print(f"   ✓ Content extracted: {len(content)} characters")
    
    # Split content into chunks
    chunks = split_text_langchain(content, chunk_size, chunk_overlap)
    
    # Create rows for each chunk
    rows = []
    for idx, chunk in enumerate(chunks, 1):
        rows.append({
            'title': title,
            'category': policy_data['category'],
            'type': policy_data['type'],
            'link': url,
            'chunk_number': idx,
            'total_chunks': len(chunks),
            # 'chunk_size': len(chunk),
            'content': chunk
        })
    
    print(f"   ✓ Created {len(chunks)} chunks")
    return rows

async def retry_failed_policies(failed_policies, chunk_size=1000, chunk_overlap=100):
    """Retry fetching failed policies one by one (slower but more reliable)."""
    connector = aiohttp.TCPConnector(limit=1, ssl=False)  # One at a time
    async with aiohttp.ClientSession(connector=connector) as session:
        all_rows = []
        
        for policy in failed_policies:
            rows = await process_single_policy(session, policy, chunk_size, chunk_overlap)
            all_rows.extend(rows)
            await asyncio.sleep(1)  # Small delay between requests
        
        return all_rows

def append_to_excel(new_data, excel_file="policies_data.xlsx"):
    """Append new data to existing Excel file."""
    try:
        # Read existing data
        existing_df = pd.read_excel(excel_file, sheet_name='Policies')
        print(f"✓ Loaded existing data: {len(existing_df)} rows")
        
        # Create dataframe from new data
        new_df = pd.DataFrame(new_data)
        
        # Ensure columns match
        column_order = ['title', 'category', 'type', 'link', 'chunk_number', 'total_chunks', 'content']
        new_df = new_df[column_order]
        
        # Append new data
        combined_df = pd.concat([existing_df, new_df], ignore_index=True)
        
        # Save back to Excel with formatting
        with pd.ExcelWriter(excel_file, engine='openpyxl') as writer:
            combined_df.to_excel(writer, sheet_name='Policies', index=False)
            
            workbook = writer.book
            worksheet = writer.sheets['Policies']
            
            # Adjust column widths
            worksheet.column_dimensions['A'].width = 30
            worksheet.column_dimensions['B'].width = 20
            worksheet.column_dimensions['C'].width = 15
            worksheet.column_dimensions['D'].width = 50
            worksheet.column_dimensions['E'].width = 12
            worksheet.column_dimensions['F'].width = 12
            # worksheet.column_dimensions['G'].width = 12
            worksheet.column_dimensions['G'].width = 80
            
            # Wrap text in content column
            from openpyxl.styles import Alignment
            for row in worksheet.iter_rows(min_row=2, max_row=worksheet.max_row, min_col=8, max_col=8):
                for cell in row:
                    cell.alignment = Alignment(wrap_text=True, vertical='top')
        
        print(f"✓ Data appended successfully!")
        print(f"✓ Previous rows: {len(existing_df)}")
        print(f"✓ New rows added: {len(new_df)}")
        print(f"✓ Total rows now: {len(combined_df)}")
        
        return combined_df
        
    except FileNotFoundError:
        print(f" Excel file not found: {excel_file}")
        print("Creating new file instead...")
        new_df = pd.DataFrame(new_data)
        column_order = ['title', 'category', 'type', 'link', 'chunk_number', 'total_chunks', 'chunk_size', 'content']
        new_df = new_df[column_order]
        new_df.to_excel(excel_file, sheet_name='Policies', index=False)
        return new_df
    except Exception as e:
        print(f" Error appending to Excel: {e}")
        return None

async def retry_main(failed_urls, excel_file="policies_data.xlsx", chunk_size=1000, chunk_overlap=100):
    """Main function to retry failed policies."""
    print("="*80)
    print("RETRY FAILED POLICIES")
    print("="*80)
    
    # Create policy list from failed URLs
    failed_policies = []
    for item in failed_urls:
        if isinstance(item, dict):
            failed_policies.append(item)
        elif isinstance(item, str):
            # If just URL string, create a basic policy dict
            failed_policies.append({
                'title': item.split('/')[-2].replace('-', ' ').title(),
                'category': 'Unknown',
                'type': 'Policy',
                'link': item
            })
    
    print(f"\nRetrying {len(failed_policies)} failed policies...")
    print(f"Using slower sequential fetching for reliability\n")
    
    start_time = datetime.now()
    new_data = await retry_failed_policies(failed_policies, chunk_size, chunk_overlap)
    end_time = datetime.now()
    
    if new_data:
        df = append_to_excel(new_data, excel_file)
        
        print(f"\n{'='*80}")
        print(f"✓ Retry complete!")
        print(f"✓ Time taken: {(end_time - start_time).total_seconds():.2f} seconds")
        print(f"✓ Successfully processed: {len([p for p in failed_policies if any(d['title'] == p['title'] for d in new_data)])} policies")
        print(f"{'='*80}")
        
        return df
    else:
        print("\n No data to append!")
        return None

# Usage in Jupyter:
# Define your failed policies
failed_policies = [
    {"title": "Guidelines for Implementation of the Plan as Related to 4-H Clubs and Florida Association for Family and Community Education Clubs", "category": "Academic Affairs", "type": "Regulation", "link": "https://policy.ufl.edu/regulation/6-021/"},
    {"title": "University of Florida Investment Policy", "category": "Finance", "type": "Policy", "link": "https://policy.ufl.edu/policy/university-of-florida-investment-policy/"},
    {"title": "Academic Personnel Employment Plan: Academic Appointments, Types of Appointments, Appointment Status Modifier, and Academic-Administrative Classification Titles", "category": "Human Resources", "type": "Regulation", "link": "https://policy.ufl.edu/regulation/7-003/"},
    {"title": "Essential Employees", "category": "Human Resources", "type": "Policy", "link": "https://policy.ufl.edu/policy/essential-employees/"},
    {"title": "Furlough Policy (includes UFF Faculty MOU)", "category": "Human Resources", "type": "Policy", "link": "https://policy.ufl.edu/policy/furlough-policy-includes-uff-faculty-mou/"},
    {"title": "Modified Duty", "category": "Human Resources", "type": "Policy", "link": "https://policy.ufl.edu/policy/modified-duty/"},
    {"title": "Office of Academic Affairs Requirements", "category": "Academic Affairs", "type": "Regulation", "link": "https://policy.ufl.edu/regulation/7-0012/"},
    {"title": "Complaints Against Faculty Members", "category": "Academic Affairs", "type": "Regulation", "link": "https://policy.ufl.edu/regulation/7-036/"},
    {"title": "Non-Commercial Aircraft Use", "category": "Operations", "type": "Policy", "link": "https://policy.ufl.edu/policy/non-commercial-aircraft-use/"},
]

# Run the retry
df = await retry_main(failed_policies, excel_file="policies_data.xlsx", chunk_size=1000, chunk_overlap=100)
```
